```python
paradoxo = input("O mal existe?")

if paradoxo == "Sim":
    paradoxo2 = input("Deus sabe que o mal existe?")

    if paradoxo2 == "Sim":
        paradoxo3 = input("Deus pode acabar com o mal?")

        if paradoxo3 == "Sim":
            paradoxo4 = input("Deus quer acabar com o mal?")

            if paradoxo4 == "Sim":
                print("Então por que o mal existe?")
            
            elif paradoxo4 == "Não":
                print("Então ele não é bom.")

            else:
                print("Resposta inválida.")
        
        elif paradoxo3 == "Não":
            print("Então ele não é onipotente.")

        else:
            print("Resposta inválida.")
    
    elif paradoxo2 == "Não":
        print("Então ele não é onisciente.")

    else:
        print("Resposta inválida.")
else:
    print("Não, o mal existe.")

## This is an example of nested conditional statements, where if, elif, or else are dependent of each other and nested inside
```

```python
num1 = float(input("Insira o primeiro número:"))
num2 = float(input("Insira o segundo número:"))
operador = input("Insira o operador (+, -, /, *, %, **):")

def calculo(num1, num2, operador):
    if operador == "+":
        return num1 + num2
    elif operador == "-":
        return num1 - num2
    elif operador == "/":
        return num1 / num2
    elif operador == "*":
        return num1 * num2
    elif operador == "%":
        return num1 % num2
    elif operador == "**":
        return num1 ** num2
    else:
        return None

try:
    resultado = calculo(num1, num2, operador)
    if resultado is not None:
        print(f"O resultado do cálculo {num1} {operador} {num2} é: {resultado}")
    else:
        print("Operador inválido.")
except ZeroDivisionError:
    print(f"Não é possível divisão ou módulo por zero.")

## Example of a simple calculator using try... except, if, elif and else, also using a function
```

```python
for x in range(4):
    for y in range(3):
        print(f"This is x: {x}; This is y: {y}")

## This is an example of nested loops, which are loops inside another loop

dicionario = {"amor": ["afeto", "carinho"],
             "legal": ["bom", "justo"]}

for palavra, sinonimos in dicionario.items():
    print(f"Palavra: {palavra}")
    for count, sinonimo in enumerate(sinonimos, start=1):
        print(f"Sinônimo {count}: {sinonimo}")

## This is another example using a dictionary and the items() method, which returns the key and its values
## The palavra and sinonimos are for both the key and value respectively
## count and sinonimo are used for the enumarate() function, which enumerates a list of objects like "(0, obj1), (1, obj2)"
## start=1 is used to make the enumeration start by 1 and not 0
```

```python
x = int(input("Insira o limite de até que número somar:"))
soma = 0
contar = 1

if x <= 0:
    print("Insira um número positivo.")
else:
    while contar <= x:
        soma += contar
        contar += 1
    print(f"A soma dos {x} primeiros números naturais é: {soma}")

## Sum of the first x natural numbers using a while loop

x = int(input("Insira o limite de até que número somar (for):"))
soma = 0

for contar in range(1, x + 1):
    soma += contar

print(f"A soma dos {x} primeiros números naturais é: {soma}")

## Another sum of the first x natural numbers using a for loop
```

```python
primos = range(1, 101)

for primo in primos:
    limite = range(2, primo)
    if primo > 1:
        for numero in limite:
            if primo % numero == 0:
                break
        else:
            print(primo)

## This calculates the prime numbers between 1 and 100
## The else along with for only executes when for executes without activating the break statement
```
